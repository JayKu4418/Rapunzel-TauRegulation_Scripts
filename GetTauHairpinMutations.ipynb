# Obtain mutations and common SNPs around MAPT hairpin

This script will obtain mutations and common SNPs found around MAPT hairpin depending on the coordinates given

```python
import pandas as pd
```

```python
# file name to append and start and end coordinates to zoom in on around MAPT hairpin
fileAppendName="_ForLaederachPrimerSet" 
#startCoord=46010377
startCoord=46010343
#endCoord=46010427
endCoord=46010543
```

```python
# Read in mutation data for HGMD and ClinVar
allmuts = pd.read_csv("../data/CombinedMutations_HGMDandClinVar.bed",sep="\t",header=None)
allmuts.head()
```

```python
# Get muts around MAPT hairpin
tau_hairpin_muts = allmuts[((allmuts[0]=="chr17")&(allmuts[1]>=startCoord)&(allmuts[2]<=endCoord))]
tau_hairpin_muts.to_csv("../processed_data/MutationsAroundTauHairpin"+fileAppendName+".bed",sep="\t",header=False,index=False)
tau_hairpin_muts.head()
```

```python
# Read in common SNPs data
commonSNPs = pd.read_csv("../data/commonSNPs_all_processed.bed",sep="\t",header=None)
commonSNPs.head()
```

```python
# Get commonSNPs around MAPT hairpin
tau_hairpin_SNPs = commonSNPs[((commonSNPs[0]==17)&(commonSNPs[1]>=startCoord)&(commonSNPs[2]<=endCoord))]
tau_hairpin_SNPs.head()
```

```python
# Add chr to 17 so that it will be chr17
tau_hairpin_SNPs.loc[:,0] = 'chr' + tau_hairpin_SNPs.loc[:,0].astype(str)
tau_hairpin_SNPs.head()
```

```python
# Write common snps around MAPT hairpin 
tau_hairpin_SNPs.to_csv("../processed_data/CommonSNPsAroundTauHairpin"+fileAppendName+".bed",sep="\t",header=False,index=False)
```

```python
%%script bash
# Concatenate mut and common snps around MAPT hairpin files
cat ../processed_data/MutationsAroundTauHairpin_ForLaederachPrimerSet.bed ../processed_data/CommonSNPsAroundTauHairpin_ForLaederachPrimerSet.bed > ../processed_data/MutsAndSNPsAroundTauHairpin_ForLaederachPrimerSet.bed
```
